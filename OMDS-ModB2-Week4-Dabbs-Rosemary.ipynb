# 2. Weekly graph question

Below are a histogram and boxplot representation of the same data. A pharmacy is keeping a record of the prices of the drugs that it sells, and an administrator wants to know how much the more expensive drugs tend to cost, in the context of the other prices.

Please write a short explanation of the pros and cons of these two representations. Which would you choose? How would you modify the formatting, if at all, to make it more visually interesting, clear, or informative?

```python
import numpy as np
import pandas as pd

np.random.seed(0)
num_data = 100
data = np.exp(np.random.uniform(size = num_data) * 4)
df = pd.DataFrame(data.T, columns = ["data"])
```

```python
print("The 75th percentile is:", df.quantile(q = 0.75))
df.plot.hist()
None
```

```python
df.plot.box()
None
```

### Pros and Cons
Pros:
- Both represent drug price distribution and can give insight to answer the given question

Cons:
- The label "data" is ambiguous.
- There are no titles.
- The units are not shown, is it dollars or dollars in thousands?
- The legend is not adding value and is redundant

### How to Improve the Charts

For the histogram and box plot, I would add specific labels and a title that clearly specifies what is being measured including units. For both plots, I might highlight the more expensive drugs, bins or datapoints, to bring attention to them since that is the focus of the administrator's question. This will make it clear to the viewer what is important. For the histogram, I would highlight how the fregency of more expensive drugs is less common by adding an overlaying curve that shows that as prices go up frequency goes down. Also, the label on the y-axis is redunant since this is a histogram, I would only label the x-axis. For the box plot, I would update the formatting of the plot to make it more visually appealing and add labels for the quantiles.

### What Graph to Choose
From these observations, I think the best choice for this particular problem is the box plot. The box plot gives more insight into how much more the expensive drugs costs because the data is broken into quantiles and the outliers or "more expensive" drugs are more clearly visible.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
phishing_data = pd.read_csv('dataset_phishing.csv')
```

```python
phishing_data.describe()
```

```python
# Get all column names
print(list(phishing_data.columns))
```

```python
# Plot histogram for URL length
bins = list(range(0, 1800, 200))
phishing_data['length_url'].hist(bins=bins)
plt.title('Distribution of All Lengths of URL')
plt.xlabel('# of characters')
plt.xlim(0, 1800)
None
```

```python
# The plot is squished a bit and is hard to read at higher values
# What is the max?
phishing_data['length_url'].max()
```

```python
# Remove some outliers to explore the more common URL lengths
df_filtered = phishing_data[["length_url"]].query('length_url < 200')
```

```python
# Plot zoomed in data and change bins
bins = list(range(0, 200, 20))
df_filtered['length_url'].hist(bins=bins, edgecolor='black')
plt.title('Distribution of More Common Lengths of URL')
plt.xlabel('# of characters')
plt.xlim(0, 200)
plt.xticks(bins)
None
```

```python
# Let's look at the feature length of hostname
bins = [0, 50, 100, 150, 200, 250, 300]
phishing_data['length_hostname'].hist(bins=bins)
plt.title('Distribution of All Lengths of Hostname')
plt.xlabel('# of characters')
plt.xticks(bins)
plt.xlim(0, 300)
None
```

```python
## Let's zoom in to look at bin 0 to 50

# Filter data
hostname_length_filtered = phishing_data[["length_hostname"]].query('length_hostname < 50')

# Create histogram
bins = list(range(51))
hostname_length_filtered['length_hostname'].hist(bins=bins, edgecolor='black')
plt.title('Distribution of Common Lengths of Hostname')
plt.xlabel('# of characters')
plt.xticks(list(range(0, 51, 5)))
plt.xlim(0, 50)
None
```

```python
sns.kdeplot(phishing_data['length_url'])
plt.title('Distribution of All Lengths of URL')
plt.xlabel('# of characters')
None
```

```python
sns.kdeplot(phishing_data['length_hostname'])
plt.title('Distribution of All Lengths of Hostname')
plt.xlabel('# of characters')
None
```

```python
sns.violinplot(df_filtered)
plt.title('Distribution of Common URL Lengths')
None
```

```python
sns.violinplot(hostname_length_filtered)
plt.title('Distribution of Common Hostname Lengths')
None
```

```python
# Filter data to look at less common lengths
url_length_filtered_uncommon = phishing_data[["length_url"]].query('length_url > 200')
hostname_length_filtered_uncommon = phishing_data[["length_hostname"]].query('length_hostname > 50')
```

```python
sns.swarmplot(url_length_filtered_uncommon)
plt.title('Distribution of Less Common URL Lengths')
None
```

```python
sns.swarmplot(hostname_length_filtered_uncommon)
plt.title('Distribution of Less Common Hostname Lengths')
None
```

```python
phishing_data['status'].value_counts().plot.bar()
plt.title('Balanced Target Categories')
None
```

## Conclusions:

The distributions of the lengths for URL and Hostname are skewed very far to the left. There appears to be outliers for both features that were explored. Both distributions for length appear to follow similar distributions. There is a range that is common and then there are outliers that are typically longer pushing the distribution to the left. The range for URL length is 12 to 1641. The range for hostname length is 4 to 214. When looking at the zoomed in histograms, the distribution of the lengths is more clear. Without changing the bins, the skewed distribution would be lost. The dataset has two target variables, phishing and legitimate, and they are balanced 50/50.
