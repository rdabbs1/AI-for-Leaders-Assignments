# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

## Advantages
> **Plot 1**:  
Putting the two line graphs side-by-side shows changes over time for the two week period more clearly. This format is less clutter and easier to point out trends specific to individual weeks. Could be useful when comparing visitation numbers over a month long period (4 weeks) or when a specific week has an unusually high or low number of visitations that needs to be highlighted.

> **Plot 2**:  
With the two lines plotted together on the same plot, the viewer can compare the two weeks peak visitation days more easily and identify patterns better. This graph also shows where visitations equaled each other at the intersection point between Wednesday (3) and Thursday (4) which could be valuable information. Viewers are able to identify what days of the week are most popular quickly.


## Disadvantages
> **Plot 1**:  
This format makes direct comparison difficult and could be misleading if axis are not consistant. Adding more weeks could take up too much space and make it difficult to compare and determine what patterns are trending per week. 

> **Plot 2**:  
With the two lines plotted together on the same plot, there is not much clutter, but adding more weeks could make this graph hard to read especially if the color scheme is not distinct. Probably not the best way to highlight anomalies in the data.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
# -------------------------------
# Recreation of graph from p. 189
# -------------------------------

products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E'] # Product name list
years = ['2008', '2009', '2010', '2011', '2012', '2013', '2014'] # Year list
n_years = len(years) # Number of years

# Create data to recreate the plot
data = [
    [390, 410, 415, 390, 300, 275, 267],  # Product A
    [365, 400, 410, 380, 293, 275, 267],  # Product B
    [0, 0, 100, 190, 200, 220, 190],  # Product C
    [0, 0, 0, 180, 260, 230, 225],        # Product D
    [0, 0, 0, 0, 0, 100, 210],            # Product E
]

# Transform
data = np.array(data).T  # (years, products)

# Check shape of data
data.shape
```

```python
# Define label location and bar width
x = np.arange(len(products))
bar_width = 0.1

# Create color list: 2008 and 2009 are gray and the remaining bars are black
colors = ['gray'] * 2 + ['black'] * (n_years - 2)

fig, ax = plt.subplots(figsize=(10, 6))

for i in range(n_years):
    ax.bar(x + i * bar_width, data[i], width=bar_width, label=years[i], color=colors[i])

# Labels and title
ax.set_title('Average Retail Product Price per Year')
ax.set_xticks(x + bar_width * 3)
ax.set_xticklabels(products)
ax.legend(title='', loc='lower center', bbox_to_anchor=(0.5, -0.25), ncol=n_years)

# Add horizontal grid lines
ax.yaxis.grid(True, linestyle='-', color='gray')
ax.set_axisbelow(True)  # Make sure grid lines are behind the bars

# Format y-axis with dollar signs
ax.set_yticks(np.arange(0, 501, 100))
ax.set_yticklabels([f"${i}" for i in range(0, 501, 100)])

plt.tight_layout()
plt.show()
```
