Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")

None
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")

None
```

If I wanted to make the viewers look at the last three months first, several preattentive attributes could be used: hue, shape, enclosure, and intensity. First, color could be used to highlight the line for the last three months. Second, markers could be added for the last three months. Third, the last three months could be enclosed in a box to bring attention to those specific months. # 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
```

```python
# Create data for plot recreation
categories = ['Engine power is less than expected',
     'Tires make excessive noise while driving',
     'Engine makes abnormal/excessive noise',
     'Seat material concerns',
     'Excessive wind noise',
     'Hesitation or delay when shifting',
     'Bluetooth system has poor sound quality',
     'Steering system/wheel has too much play',
     'Bluetooth system is difficult to use',
     'Front seat audio/entertainment/navigationcontrols']

values = [12.9, 12.3, 11.6, 11.6, 11.0, 10.3, 10.0, 8.8, 8.6, 8.2]

# Create a DataFrame
df = pd.DataFrame({'Category': categories, 'Value': values})

# Sort the DataFrame by the 'Value' column in ascending order
df_sorted = df.sort_values(by='Value')

# Recreate bar plot on page 109
fig, ax = plt.subplots()
bars = ax.barh(df_sorted['Category'], df_sorted['Value'], color='black')
ax.set_title('Top 10 Design Concerns')
ax.title.set_position([-0.5, 0])

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Remove x-ticks (the tick marks along the x-axis)
ax.set_xticks([])
ax.tick_params(axis='y', which='both', length=0)
ax.bar_label(bars, labels={v for v in df_sorted['Value']}, color='white', fontweight='bold', padding=-30)

# Add unit label
ax.text(0, 9.6, 'concerns per 1,000', color='black')

None
```
