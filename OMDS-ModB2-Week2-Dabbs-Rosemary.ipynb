```python
# Summarize the datasets
phishing_data.info()
```

```python
phishing_data.describe()
```

```python
# Check for duplicates
duplicates = phishing_data.duplicated().sum()

if duplicates == 0:
    print('No duplicates found.')
else:
    print(f'There are {duplicates} duplicates in the dataset.')
```

```python
# Check for null values
nulls = phishing_data.isnull().sum().sum()

if nulls == 0:
    print('No missing values found.')
else:
    print(f'There are {nulls} missing data fields in the dataset.')
```

```python
# Calculate mean, median, and mode
means = phishing_data.mean(numeric_only=True)
print('Means for the Data by Column')
print(means)
```

```python
medians = phishing_data.median(numeric_only=True)
print('Medians for the Data by Column')
print(medians)
```

```python
modes = phishing_data.mode(numeric_only=True).iloc[0]
print('Modes for the Data by Column')
print(modes)
```

```python
# Encode status column
one_hot = pd.get_dummies(phishing_data["status"])
df_one_hot = phishing_data.join(one_hot).copy()

# Convert to integers
df_one_hot[['legitimate','phishing']] = df_one_hot[['legitimate','phishing']].astype(int)

# Check to see if encoding worked
df_one_hot.iloc[0:5]
```

```python
import matplotlib.pyplot as plt
```

```python
# Create data for chart recreation
fake_data = {'Ticket Volume Received': [160, 184, 241, 149, 180, 161, 132, 202, 160, 139, 149, 177], 
             'Ticket Volume Processed': [160, 184, 237, 148, 181, 150, 123, 156, 126, 104, 124, 140]}

df = pd.DataFrame(fake_data)
df = df.rename(index={0: 'January', 
                 1: 'Febuary', 
                 2: 'March', 
                 3: 'April', 
                 4: 'May', 
                 5: 'June', 
                 6: 'July', 
                 7: 'August', 
                 8: 'September', 
                 9: 'October', 
                 10: 'November', 
                 11: 'December'}).copy()
```

```python
# Reset index
df = df.reset_index().copy()
```

```python
# Recreate line plot from introduction page 3
ax = df.plot(kind="line", legend=False, color=['gray', 'blue'])
ax.set_ylabel('Number of Tickets', loc='top')
ax.set_title('Ticket Volume Over Time')

# Fix x-axis labels
labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
ax.set_xticks(range(12))
ax.set_xticklabels(labels)

# Remove tick marks from x-axis
ax.tick_params(axis='x', which='both', length=0)

# Fix y-axis scale
ax.set_ylim(0, 300)
ax.set_yticks(range(0, 301, 50))

# Annotate lines and points
plt.annotate('Received', xy=(11, 177), xytext=(11.3, 177), color='gray')
plt.annotate('Processed', xy=(11, 140), xytext=(11.3, 140), color='blue')

# Remove lines around chart
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Apply marker style to points from August to December
df.iloc[7::].plot(kind="line", marker='o', legend=False, color=['gray', 'blue'], ax=ax)

# Add vertical line at May
ax.axvline(x=4, color='gray', linewidth=0.7, ymax=0.74)

# Add text to graph
text1 = "2 Employees Quit in May."
text2 = "We nearly kept up with incoming volume"
text3 = "in the following two months, but fell behind with the increase in Aug \nand haven't been able to catch up since."
plt.text(3.1, 255, text1, fontweight='bold')
plt.text(8, 255, text2)
plt.text(3.1, 229, text3)

None
```
